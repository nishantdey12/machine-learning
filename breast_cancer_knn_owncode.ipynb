k nearest neighbours is based on closeness. for k=3, it means the the 3 points which are closest together. and closeness is measured by eucleidian distance

```python
dataset={'k':[[1,2],[2,3],[3,1]],'r':[[6,5],[7,7],[8,6]]}
#two classes (k, r)and their features (everything else)
new_features=[5,7]
```

```python
for i in dataset:
        for j in dataset[i]:
                plt.scatter(j[0],j[1], color=i)
plt.scatter(new_features[0], new_features[1], c='green')
plt.show()
```

```python
def k_nearest_neighbours(data, predict, k=3):
        if(len(data)>=k):
                warnings.warn('k is set to a value less than total voting groups')
        distances=[]
        for group in data:
                for features in data[group]:
                        # eucleidian_distance=np.sqrt( np.sum( ( np.array(features)-np.array(predict) ) **2 ))
                        eucleidian_distance=np.linalg.norm(np.array(features)-np.array(predict))
                        distances.append([eucleidian_distance, group])
        votes=[i[1]  for i in sorted(distances)[:k]]
        print(Counter(votes).most_common(1))
        vote_result=Counter(votes).most_common(1)[0][0]
        # for i in sorted(distances)[:k] :
        #         i[1]
        return vote_result


result=k_nearest_neighbours(dataset, new_features, k=3)
print(result)
```
