```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import neighbors, preprocessing
from sklearn.linear_model import LinearRegression
import pandas as pd
# import matplotlib.pyplot as plt
#cross validation is train
```

```python
df=pd.read_csv('D:\\Jupyter Practice\\sentdex\\k nearest neighbours\\breast-cancer-wisconsin.data')
df.replace('?',-999999, inplace=True)

# df.show()
df.drop(['Sample code number'], 1, inplace=True)
df.plot.area( subplots=True)  #this is used to plot each axis separately
```

```python
x=np.array(df.drop(['Class'],1))
y=np.array(df['Class'])
x_train, x_test, y_train,y_test=train_test_split(x, y, test_size=0.2, random_state=100)
clf=neighbors.KNeighborsClassifier()
clf.fit(x_train, y_train)
accuracy=clf.score(x_test, y_test)
print(accuracy)
```

```python
clf_1=LinearRegression()
clf_1.fit(x_train, y_train)
accuracy_1=clf_1.score(x_test, y_test)
print(accuracy_1)
```

```python
example_measures=np.asarray([[9,6,9,2,10,6,2,9,10]])
example_measures.reshape(1,-1)
prediction=clf.predict(example_measures)
print(prediction)
```
